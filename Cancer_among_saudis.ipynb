```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv('Ranking of the Most Common Cancers Among Saudi 1.csv')
                 
```

```python
df
```

```python
df = df.iloc[1:]
```

```python
df
```

```python
df_mean = df.groupby('Sites')['Count'].mean().sort_values(ascending=False)
```

```python
df_mean
```

```python
df_mean.plot(kind='barh')
```

```python
sites = df_mean.index
counts = df_mean.values


bar_colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 
              'tab:purple', 'tab:brown', 'tab:pink', 'tab:gray', 
              'tab:olive', 'tab:cyan']

bar_labels = sites

fig, ax = plt.subplots(figsize=(10,6))
bars = ax.bar(sites, counts, color=bar_colors, label=bar_labels)

ax.set_ylabel('Count')
ax.set_title('Ranking of the Most Common Cancers Among Saudi Nationals 2022')
ax.legend(title='Cancer Site', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.xticks(rotation=45)
plt.tight_layout()
sns.set_style('darkgrid')
plt.savefig('Cancer cases by sites.png',dpi=300)
plt.show()
```
